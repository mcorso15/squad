## Task 2 - Visualizing the data
### Task 2.1

```python
# read in data and create variables
path1 = ('../data/DAS/Global_DAS_1078DT_25PR_14GL_5DEC_2023-02-17T164133Z.h5')
strain_rate, time = strain_rate_and_mountain_time(path1)
n_time, n_channels = strain_rate.shape
channels = np.arange(n_channels)
```

```python
# plot colormap
vmin = -5e5
vmax = 5e5
dd = 5
cs = plt.pcolormesh(channels[::dd], time[::dd], strain_rate[::dd,::dd], vmin=vmin, vmax=vmax, cmap='RdBu_r')
cbar = plt.colorbar(cs)
cbar.set_label('Strain Rate [s$^{-1}$]', fontsize=12)
plt.xlabel('Channel')
plt.ylabel('Time')
plt.title(time[0].date())
```
